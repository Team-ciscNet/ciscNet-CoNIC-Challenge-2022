```python
import joblib

import numpy as np
import matplotlib.pyplot as plt
```

```python
STATS_PATHS = [
    '/srv/scherr/projects/conic_challenge/hover_net/exp_output/local/models/isbi_03/00/model/00/stats.dat',
    '/srv/scherr/projects/conic_challenge/hover_net/exp_output/local/models/isbi_03/00/model/01/stats.dat'
]
NUM_CLASSES = 7
```

```python
plt.figure(dpi=150)
for i, stats_path in enumerate(STATS_PATHS):
    stats = joblib.load(stats_path)
    dice = np.zeros(len(stats))
    for j in range(NUM_CLASSES):
        class_metrics = [y[f'infer-valid-tp_dice_{j}'] for _, y in stats.items()]
        dice += class_metrics
    dice /= NUM_CLASSES
    best_epoch = np.argmax(dice) + 1
    print(f'Best score at epoch {best_epoch} for {stats_path}')
    plt.plot(dice, label=i)
plt.legend()
plt.show()
plt.close()
```
